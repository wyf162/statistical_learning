#  封装

```python
# 数据线性可分，二分类数据
class Model:
    def __init__(self):
        self.w = 0
        self.b = 0
        self.learning_rate = 0
        # 误分类点集合的大小，用于可视化训练的过程
        self.error_set_length = []
        # self.data = data

    def sign(self, x, w, b):
        y = np.dot(x, w) + b
        return y

    # 定义训练函数
    def train(self,X_train, Y_train, learning_rate):
        self.w = np.ones(X_train.shape[1])
        self.learning_rate = learning_rate
        # 新建误分类点集合
        error_set=set()
        # 构建误分类点集合
        for i in range(X_train.shape[0]):
            x = X_train[i]
            y = Y_train[i]
            if y * sign(x, self.w, self.b) <= 0:
                error_set.add(i)
        self.error_set_length.append(len(error_set))
        while len(error_set):
            i = error_set.pop()
            x = X_train[i]
            y = Y_train[i]
            self.w = self.w + self.learning_rate * y * x
            self.b = self.b + self.learning_rate * y
            # 构建误分点集合
            for i in range(X_train.shape[0]):
                x = X_train[i]
                y = Y_train[i]
                if y * sign(x, self.w, self.b) <= 0:
                    error_set.add(i)
            self.error_set_length.append(len(error_set))    
        return 'Perceptron Model!'
 

    def draw_plot(self):
        x_axis = [i for i in range(len(self.error_set_length))]
        y_axis = self.error_set_length
        plt.plot(x_axis,y_axis)
```

```python
perceptron = Model()
perceptron.train(X,Y,0.1)
perceptron.draw_plot()
```
